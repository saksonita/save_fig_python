```python
## Save Figure on Each Date
from datetime import date, timedelta

def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = df.index.min()
end_date = df.index.max()
print (start_date.strftime('%Y-%m-%d'))

for single_date in daterange(start_date, end_date):
    single_date = single_date.strftime('%Y-%m-%d')
    new_filtered = df.loc[single_date]
    new_filtered
    fig = px.line(new_filtered,x=new_filtered.index, y='_value', color='_field', width =2000, height=900, template='plotly_dark')
    fig.update_layout({
                   "xaxis": {"title":"Date and Time"},
                   "showlegend": True})
    fig.write_image("save_fig/"+single_date+".png", engine="auto")
```
